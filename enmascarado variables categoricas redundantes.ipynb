```
predictions = lm.predict(X)
```

```
df_new["predictions"] = predictions
```

```
df_new.head()
```

suma de los caudrados de las diferencias

```
import numpy as np
```

suma de los cuadrados de las diferencias

```
SSD = np.sum((df_new["predictions"] - df_new["Total Spend"])**2)
```

```
SSD
```

Desviacion tipica de los residuos

```
RSE = np.sqrt(SSD / (len(df_new) - len(feature_cols)-1))
```

```
RSE
```

Desviacion de 803 pesos por arriba y 803 pesos por abajo

Promedio de las ventas

```
sales_mean = np.mean(df_new["Total Spend"])
```

```
sales_mean
```

```
error = RSE / sales_mean * 100
```

```
column_names = df.columns.values.tolist()
df_new = df[column_names].join(dummy_gender)
column_names = df_new.columns.values.tolist()
df_new = df_new[column_names].join(dummy_city_tier)
df_new.head()
```
